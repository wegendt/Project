#  Elliptisch polarisiertes Licht 

Beim elliptisch polarisiertem Licht ändert sich der Betrag des E-Vektors mit der Zeit in Form einer Ellipse. 

$\frac{x^2}{a^2}+\frac{y^2}{b^2}=1$    


Wähle für den Plot z.B. 

a=1 

b=3

Plotte in Polarkoordinaten: 

$x=r cos (\phi)$

$y=r sin (\phi)$

$r(\phi)$, wobei $r(\phi)$ für den Betrag des E-Vektors steht $|E(t)|$.

$r=\frac{a b}{\sqrt{b^2 \cos^2 \phi+a^2 \sin^2 \phi}}$


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

a=1
b=3

phi_gra = np.arange(0, 360, 1) #Winkel in Grad in 1° Schritten: array([0, 1, ..., 358, 359])
phi = phi_gra * np.pi /180 #Winkel in Rad: array([0. , 0.01745329, ... 6.24827872, 6.26573201])

off=25*np.pi/180  # Offset Winkel, weil - warum eigentlich?

r=a*b*((b*np.cos(phi + off))**2 + (a*np.sin(phi+off))**2)**(-0.5)

fig, ax = plt.subplots()
ax = plt.subplot(111, projection='polar')
ax.plot(phi, r)
ax.set_rlabel_position(90)
ax.grid(linestyle='--')
plt.show()
```

# Gesetz von Malus
Das Licht passiert den Analisator, der laut Gesetz von Malus nur die Intensität $(E*cos(\lambda))^2$ durchlässt. 


```python
lamb_gra_20 = 20                        # zum Zeichnen wählen wir irgendein Lambda, z.B. lambda =20°
lamb_20 = lamb_gra_20 * np.pi /180

lamb_gra_40 = 40          # und noch eins, z.B. lambda =40° 
lamb_40 = lamb_gra_40 * np.pi /180

lamb_gra_60 = 60        # und noch eins, z.B. lambda =60° 
lamb_60 = lamb_gra_60 * np.pi /180

proj_20 = ( r*np.cos(lamb_20 - phi) )**2    
proj_40 = ( r*np.cos(lamb_40 - phi) )**2
proj_60 = ( r*np.cos(lamb_60 - phi) )**2

fig, ax = plt.subplots()
ax = plt.subplot(111, projection='polar')
ax.plot(phi, proj_20)
ax.plot(phi, proj_40)
ax.plot(phi, proj_60)
ax.set_rlabel_position(90)
ax.grid(linestyle='--')

plt.show()
```

# Überlagerung des Lichts bei jeder einzelnen Analysatoreinstellung
Bei einer Analisatorstellung wird über die Zeit addiert, der E-Vektor rotiert und ändert ellipsenförgig seine Größe. Diese wird auf die Richtung des Analysators projiziert, aber letztendlich gelangt alles Licht in die Diode. Die Intensität des Lichts über alle Winkel bei einer Analysatorstellung wird addiert.


```python
lamb_gra=0
x=np.zeros((360,1))
delta=1
for i in range (0, 360):
    lamb_gra = lamb_gra + delta
    lamb = lamb_gra * np.pi /180
    proj = ( r*np.cos(lamb - phi) )**2
    k=sum(proj)
    x[i]=k

```

```python
fig, ax = plt.subplots()
ax = plt.subplot(111, projection='polar')
ax.plot(phi, x)
ax.set_rlabel_position(90)
ax.grid(linestyle='--')

plt.show()
```
