## Polarisation 

### 1. Gesetz von Malus

Die Messwerte aus dem Versuch sollen in eine Tabelle eingegeben werden und im Diagramm zusamen mit den berechneten Werten geplottet werden. 
$I(\alpha)= I(0°)\cdot cos^2(\alpha)$ 




```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

alpha = np.arange(0, 100, 10) #Winkel in Grad in 10° Schritten: array([0, 10, ...,90])

Winkel=np.arange(0, 91, 1) #Winkel in Grad in 1° Schritten: array([0, 10, ...,90])
```

Bitte in der nächsten Zelle die gemessenen Werte eingeben. 
Zahlenformat: Trennung mit Punkt. z.B. $\pi$=3.14

```python
#Gemessene Werte für das Gesetz von Malus 
I_gemessen_0=6.1
I_gemessen_10=5.9
I_gemessen_20=5.47
I_gemessen_30=4.74
I_gemessen_40=3.76
I_gemessen_50=2.65
I_gemessen_60=1.71
I_gemessen_70=0.82
I_gemessen_80=0.23
I_gemessen_90=0
```

In der nächsten Zelle wird I_gemessen und I_berechnet geplottet.

```python
I_gemessen=[I_gemessen_0, I_gemessen_10, I_gemessen_20, I_gemessen_30, I_gemessen_40, I_gemessen_50, I_gemessen_60, I_gemessen_70, I_gemessen_80, I_gemessen_90]

I_berechnet=I_gemessen_0* np.cos(Winkel*np.pi/180)**2

fig, ax = plt.subplots()
ax.plot(alpha, I_gemessen, linestyle='-',color='C4', label='I_gemessen', marker='s');
ax.plot(Winkel, I_berechnet, linestyle='-',color='C1', label='I_gerechnet');
ax.set_xlabel('Winkel alpha/°')
ax.legend(loc="upper right")
ax.set_ylabel('Intensität in Skalenteilen')
ax.set_title('Überprüfung des Gesetzes von Malus')
plt.show()
```

### 2. Zirkular polarisiertes Licht

Um zirkular polarisiertes Licht zu erhalten, braucht man ein $\lambda$ /4 Plättchen, das von zwei aufeinander sekrecht stehenden linear polarisierten Strahlen durchleuchtet wird, deren Amplitude gleich groß ist. Das erreicht man, indem das $\lambda$ /4 Plättchen in einen Winkel von 45° zum linearen Polarisator gestellt wird.   

Im Video in der nächsten Zelle wird gezeigt, wie sich zwei Wellen, die einen Phasenunterschied $\delta=\lambda/4$ haben, zu einer zirkular polarisierten Welle addieren.

Bitte in der nächsten Zelle die gemessenen Werte des $\textbf{zirkular}$ polarisierten Lichts eingeben (Analysatorstellung -90° bis +90°). 

```python
#Gemessene Werte für zirkular polarisiertes Licht
I_zirk_minus90=2.9
I_zirk_minus80=2.9
I_zirk_minus70=2.87
I_zirk_minus60=2.83
I_zirk_minus50=2.77
I_zirk_minus40=2.69
I_zirk_minus30=2.65
I_zirk_minus20=2.63
I_zirk_minus10=2.6
I_zirk_0=2.6
I_zirk_plus10=2.64
I_zirk_plus20=2.69
I_zirk_plus30=2.69
I_zirk_plus40=2.69
I_zirk_plus50=2.76
I_zirk_plus60=2.8
I_zirk_plus70=2.87
I_zirk_plus80=2.89
I_zirk_plus90=2.88

I_zirkular=[I_zirk_minus90, I_zirk_minus80, I_zirk_minus70, I_zirk_minus60, I_zirk_minus50, I_zirk_minus40, I_zirk_minus30, I_zirk_minus20, I_zirk_minus10, I_zirk_0, I_zirk_plus10, I_zirk_plus20, I_zirk_plus30, I_zirk_plus40, I_zirk_plus50, I_zirk_plus60, I_zirk_plus70, I_zirk_plus80, I_zirk_plus90]
```

```python
alpha_zirk=np.arange(0, 181, 10)
grad_zirk_plus=(alpha_zirk-90) * np.pi /180
grad_zirk_minus=(-alpha_zirk-90) * np.pi /180

fig, ax = plt.subplots()
ax = plt.subplot(111, projection='polar')
ax.plot(grad_zirk_plus, I_zirkular, color='blue', marker='s', markersize=4)
ax.plot(grad_zirk_minus, I_zirkular, color='blue', marker='s', markersize=4)
ax.set_rlabel_position(90)
#ax.set_rlim(0,5)
ax.grid(linestyle='--')
plt.show()
```
